```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\HP\\Downloads\\titanic.csv')
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
plt.matshow(correlation_matrix, fignum=1)
plt.xticks(range(data.shape[1]), data.columns, rotation=90)
plt.yticks(range(data.shape[1]), data.columns)
plt.colorbar()
plt.title('Feature Correlation Matrix', pad=20)
plt.show()
```

```python
X = data.drop('Survived', axis=1)
y = data['Survived']
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Fare'] = data['Fare'].fillna(data['Fare'].mean())
```

```python

X = (X - X.mean()) / X.std()
```

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.loss_history = []

    def compute_loss(self, y, y_predicted):
        # Binary cross-entropy loss
        loss = -np.mean(y * np.log(y_predicted + 1e-15) + (1 - y) * np.log(1 - y_predicted + 1e-15))
        return loss

    def fit(self, X, y):
        self.m, self.n = X.shape
        self.weights = np.zeros(self.n)
        self.bias = 0

        for _ in range(self.num_iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = sigmoid(linear_model)

            # Calculate and store the loss
            loss = self.compute_loss(y, y_predicted)
            self.loss_history.append(loss)

            # Gradient descent
            dw = (1 / self.m) * np.dot(X.T, (y_predicted - y))
            db = (1 / self.m) * np.sum(y_predicted - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = sigmoid(linear_model)
        y_predicted_cls = [1 if i > 0.5 else 0 for i in y_predicted]
        return np.array(y_predicted_cls)

# Evaluation metrics
def calculate_metrics(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    TN = np.sum((y_true == 0) & (y_pred == 0))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    
    accuracy = (TP + TN) / len(y_true)
    precision = TP / (TP + FP) if (TP + FP) != 0 else 0
    recall = TP / (TP + FN) if (TP + FN) != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    return accuracy, precision, recall, f1_score
```

```python
model = LogisticRegression(learning_rate=0.01, num_iterations=1000)
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Calculate metrics
accuracy, precision, recall, f1 = calculate_metrics(y, predictions)
print(f'Accuracy: {accuracy * 100:.2f}%')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')

# Plot loss function vs iteration
plt.plot(range(model.num_iterations), model.loss_history)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss Function vs. Iteration')
plt.show()
```
